## 1) Processing data on location of the countries

```python
def transfo(a): 
    m = a.split(",")
    if len(m) == 1:
        return int(a)
    e = len(m[1])
    return int(m[0])+int(m[1])/10**e
```

```python
L=[]
with open('Data_localisation/Urban.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')
    for row in (spamreader):
        l = [row[0]]
        for i in range(4,len(row)):
            if row[i] != '':
                l.append(transfo(row[i]))
        L.append(l)
```

```python
L2 = [L[0]]
for i in range(len(L[1:])):
    if len(L[1:][i]) == 61:
        L2.append(L[1:][i])
```

```python
df = pd.DataFrame(L2[1:],columns=L2[0][:-1])
```

```python
def transpose(L):
    l = []
    for i in range(61):
        u = []
        for j in range(len(L)):
            u.append(L[j][i])
        l.append(u)
    return l
```

```python
L1 = transpose(L2)
```

```python
df1 = pd.DataFrame(L1[1:],columns=L1[0])
```

# Mix all data together

```python
urban = pd.read_csv('./cleaned_data/urbanization.csv')
urban.drop(columns='Unnamed: 0', inplace=True)
```

```python
urban.rename(columns={'Country Name': 'dates'}, inplace = True)
```

```python
urban_ = urban[['dates','American Samoa', 'Canada', 'France', 'United Kingdom']].rename(columns={'American Samoa': 'US'})
urban_ = urban_[32:60]
urban_
```

```python
gdp = pd.read_csv('./cleaned_data/gdp.csv')
gdp = gdp[['dates', 'United States', 'Canada', 'France', 'United Kingdom']].rename(columns={'United States': 'US'})

gdp_ = gdp[32:60]
gdp_.head()
```

```python
oil = pd.read_csv('./cleaned_data/oil_consumption.csv')
oil_ = oil[27:55]
oil_.head()
```

```python
ind_ = pd.read_csv('./cleaned_data/industry_sector.csv')
ind_.head()
```

```python
ser_ = pd.read_csv('./cleaned_data/services_sector.csv')
ser_.head()
```

```python
car = pd.read_csv('./cleaned_data/Nb_Cars.csv', delimiter=';')
car = car[['dates', 'USA', 'Canada', 'France', 'UK']].rename(columns={'USA':'US', 'UK': 'United Kingdom'})
car_ = car[32:60]
```

```python
# car[32:39]['Canada'] = 23637104.0
# car[32:35]['France'] = 30040000.0
# car[36:40]['France'] = 33090000.0
# car[32:40]['United Kingdom'] = 27200000.0
car_
```

```python
car_['US'] = car_['US']/1e6
car_['Canada'] = car_['Canada']/1e6
car_['France'] = car_['France']/1e6
car_['United Kingdom'] = car_['United Kingdom']/1e6
```

```python
car_
```

```python
dates = range(1992, 2020)
countries = ['US', 'Canada', 'France', 'United Kingdom']
# data_df = pd.DataFrame(columns=['dates', 'country', 'urbanization',
#                                 'gas_price', 'num_car', 'oil_consumption', 'industry', 'services'])
data_ar = []
for c in countries:
    if(c == 'US'):
        price_df = df_usa_
    if(c == 'Canada'):
        price_df = df_can_
    if(c == 'France'):
        price_df = df_fr_
    if(c == 'United Kingdom'):
        price_df = df_uk_
    for y in dates:
        urban_i = urban_[urban_['dates'] == y][c].values[0]
        gas_price_i = price_df[price_df['dates'] == y]['price'].values[0]
        gdp_i = gdp_[gdp_['dates'] == y][c].values[0]
        num_car_i = car_[car_['dates'] == y][c].values[0]
        oil_i = oil_[oil_['dates'] == y][c].values[0]
        ind_i = ind_[ind_['dates'] == y][c].values[0]
        ser_i = ser_[ser_['dates'] == y][c].values[0]
        obs = [y, c, urban_i, gas_price_i, gdp_i, num_car_i, oil_i, ind_i, ser_i]
        data_ar.append(obs)
data_ar
```

```python
data_df = pd.DataFrame(data_ar, columns=['dates', 'country', 'urbanization',
                                'gas_price', 'gdp', 'num_car', 'oil_consumption', 'industry', 'services'])
data_df
```
